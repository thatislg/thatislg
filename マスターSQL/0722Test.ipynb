```python
# Check data shape
print(df.shape)
```

```python
#df.drop_duplicates()
df['new'] = df['odpt:startingBusstopPole'].str.replace('odpt.BusstopPole:YokohamaMunicipal.','')
show_map_df = df[['@id','geo:lat', 'geo:long', 'odpt:busNumber','new']]
show_map_df
```

```python
m = folium.Map(location=[35.388981, 139.566406],zoom_start=11)
```

```python
show_map_df.at[0,'geo:lat']
```

```python
for i in range(0,281):
    folium.Marker(
        location=[show_map_df.at[i, 'geo:lat'], show_map_df.at[i, 'geo:long']],
        popup=show_map_df.at[i,'new'],
        tooltip="Bấm vào đây xem tên điểm dừng xe bus",
        icon=folium.Icon(icon="bus"),
    ).add_to(m)
```
