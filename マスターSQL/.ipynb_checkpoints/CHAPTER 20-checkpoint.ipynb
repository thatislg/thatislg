1 まだ一度も注文されていない書籍の書籍 ID と書籍名の一覧を、結果表の見出しが「書籍 ID」、「書籍名」となるよ
うに表示しなさい。
Ứng dụng EXCEPT UNION AND INTERSECT

```python
%%sql
/*
+ Lấy thông tin:  
     mã sách: 書籍 ID
    ,tên sách: 書籍名
+ Tại bảng
    book
    ,order_detail  
+ Điều kiện: 
    Các cuốn chưa được mua lần nào   
+ Hiển thị:
    Theo trình tự lấy thông tin
*/
SELECT 
    b.book_id AS "書籍 ID"
    ,b.book_name AS "書籍名"
FROM
    -- Lấy phần khác nhau của book và order_detail
    ( SELECT book_id FROM book
      EXCEPT
      SELECT item_id FROM order_detail
    ) AS condition_book
    
    -- join lại với chính book
    INNER JOIN book AS b
        ON b.book_id = condition_book.book_id
ORDER BY 
    b.book_id
    ,b.book_name
;
```

2 会員の子供について、名前の末尾が「子」または「郎」のみのレコードを一覧表示しなさい。結果表は「会員 ID」、
「姓」、「子供名」を見出しとして表示しなさい。並び順は会員 ID の昇順、会員 ID が同一の子供については、
CHILD_NAME1、CHILD_NAME2、CHILD_NAME3 の順に表示しなさい。

```python
%%sql
/*
+ Lấy thông tin:  
     mã hội viên: 会員 ID
    ,tên họ: 姓
    ,tên con: 子供名
+ Tại bảng
    member
+ Điều kiện: 
    chữ cuối tên là 子 hoặc 郎
+ Hiển thị:
    Hiển thị tăng dần theo mã hội viên
    ,Trong trường hợp cùng mã hội viên thì hiển thị lần lượt
    theo CHILD_NAME1、CHILD_NAME2、CHILD_NAME3 
*/

WITH result AS(
                SELECT
                    member_id 
                    ,last_name 
                    ,CHILD_NAME1 AS "子供名"
                    ,1 AS order_key
                FROM member
    
                UNION 
    
                SELECT
                    member_id  
                    ,last_name 
                    ,CHILD_NAME2
                    ,2 AS order_key
                FROM member
    
                UNION 
    
                SELECT
                    member_id  
                    ,last_name 
                    ,CHILD_NAME3
                    ,3 AS order_key
                FROM member
            )
SELECT 
    member_id AS "会員 ID"
    ,last_name AS "姓"
    ,子供名
FROM 
    result
WHERE
    子供名 like '%子' OR 子供名 like '%郎'
ORDER BY 
   member_id
   ,order_key
```

3 まだ一度も注文で利用されたことのない、会員のカード情報の一覧を表示しなさい。結果表は「会員 ID」、「カード
番号」、「クレジットカード会社名」を見出しとして表示しなさい。表示順は会員 ID、カード番号とする。

```python
%%sql
/*
+ Lấy thông tin:  
     mã hội viên: 会員 ID
    ,số mã thẻ: カード 番号
    ,tên công ty thẻ: クレジットカード会社名
+ Tại bảng:
    member_card
    ,card_company
    ,order_header
+ Điều kiện: 
    Thẻ chưa được sử dụng lần nào
+ Hiển thị:
    Hiển thị tăng dần theo mã hội viên
    ,mã thẻ
*/
WITH not_use_card AS
(
    SELECT 
        member_id
        ,member_card_id
    FROM 
        member_card
    EXCEPT
    SELECT
        member_id
        ,member_card_id
    FROM
        order_header
) 
SELECT
    mc.member_id AS "会員 ID"
    ,mc.card_number AS "カード 番号"
    ,cc.card_company_name AS "クレジットカード会社名"
FROM
    not_use_card AS nuc
    INNER JOIN member_card AS mc
        ON nuc.member_id = mc.member_id AND nuc.member_card_id = mc.member_card_id
    LEFT JOIN card_company AS cc
        ON mc.card_company_id = cc.card_company_id
ORDER BY
    mc.member_id
    ,mc.card_number   
;
```

4 出版社とクレジットカード会社の全てを一覧表示しなさい。結果表は「会社名」、「所在地」、「会社種類」を見出しと
して表示しなさい。会社種類は出版社の場合は「出版社」、クレジットカード会社の場合は「カード会社」と表示する
こと。また、会社名で並び変えること。

```python
%%sql
/*
+ Lấy thông tin:  
     Tên công ty: 会社名
    ,địa chỉ: 所在地
    ,tên loại công ty: 会社種類
+ Tại bảng:
    publisher
    ,card_company
+ Điều kiện: 
    Tên tất cả các công ty làm thẻ và xuất bản
+ Hiển thị:
    tên loại công ty: 
        出版社: công ty xuất bản
        カード会社: với công ty làm thẻ
    Sắp sếp theo tên công ty
*/
WITH concat_publish_card AS(
    SELECT 
        prefecture 
        ,publisher_name AS "会社名"
        ,1 AS order_key
    FROM publisher
    UNION
    SELECT
        prefecture
        ,card_company_name AS "会社名"
        ,2 AS order_key
    FROM card_company        
)
SELECT 
    会社名
    ,prefecture AS "所在地"
    ,(CASE 
         WHEN order_key = 1 THEN '出版社'
         WHEN order_key = 2 THEN 'カード会社'
      END) AS 会社種類
FROM 
    concat_publish_card
ORDER BY 
    会社名;
```
