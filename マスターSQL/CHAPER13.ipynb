```python
%%sql
SELECT
    member_id, order_no
FROM
    order_header
WHERE 
    EXTRACT(YEAR FROM order_datetime) = '2011'
ORDER BY 
    member_id ASC,
    order_datetime DESC
;
```

```python
%%sql
SELECT
    member_id AS "会員 ID"
    ,first_name || last_name AS "氏名"
    ,blood_type AS "血液型"
    ,member_type AS "会員区分"
FROM
    member
ORDER BY 
    member_type ASC
    ,blood_type DESC
    ,member_id ASC
;
```

```python
%%sql
SELECT
    category_id AS "カテゴリ ID"
    ,ROUND(AVG(price),0) AS "平均価格"
FROM
    book
GROUP BY 
    category_id
ORDER BY 
    AVG(price) ASC
;
```

```python
%%sql
SELECT
    member_id AS "会員 ID"
    ,COUNT(member_card_id) AS "所持枚数"
FROM
    member_card
GROUP BY 
    member_id
ORDER BY
    COUNT(member_card_id)DESC
    ,member_id ASC
;
```

```python
%%sql 
SELECT
    blood_type AS "血液型"
    ,COUNT(*) AS "人数"
FROM
    member
WHERE 
    blood_type IS NOT NULL
GROUP BY 
    blood_type
HAVING 
    COUNT(*) >= 2
ORDER BY
    COUNT(*) DESC
;
```

```python
%%sql
SELECT 
    book_id AS "書籍 ID"
    ,book_name AS "書籍名"
FROM
    book
WHERE 
    publisher_id IS NOT NULL
ORDER BY
    book_id ASC
;
```

```python
%%sql
SELECT
    order_no AS "注文番号"
    ,member_id AS "会員ID"
    ,member_card_id AS "カード番号"
FROM
    order_header
WHERE   
    EXTRACT(YEAR FROM order_datetime) = '2012'
;
```

```python
%%sql
SELECT
    prefecture AS "所在地"
    ,COUNT(publisher_id) AS "出版数"
FROM
    publisher
GROUP BY 
    prefecture
HAVING
    COUNT(publisher_id) >= 1
;
```
